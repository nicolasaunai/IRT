```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

faraday_test = h5py.File("/home/debian/build/tests/faraday/faraday_test.h5")

def print_tree(name, obj):
    print(name)

faraday_test.visititems(print_tree)


Bx = faraday_test["Bx"][:]
By = faraday_test["By"][:]
Bz = faraday_test["Bz"][:]
```

## FARADAY TEST

According to Faraday's Law: 
$$ \frac{\partial B}{\partial t} = - \nabla \times E$$
We get that:
$$ B_{\rm{new}} = B_{t} - dt (\nabla \times E)$$
Then component wise:
$$ B_{\rm{new},x} = B_{t,x} - dt\left( \frac{\partial E_z}{\partial y} - \frac{\partial E_y}{\partial z}\right)$$
$$ B_{\rm{new},y} = B_{t,y} - dt\left( \frac{\partial E_x}{\partial z} - \frac{\partial E_z}{\partial x}\right)$$
$$ B_{\rm{new},z} = B_{t,z} - dt\left( \frac{\partial E_y}{\partial x} - \frac{\partial E_x}{\partial y}\right)$$
But the derivatives wrt y and z do not exist so we get simply:
$$ B_{\rm{new},x} = B_{t,x} $$
$$ B_{\rm{new},y} = B_{t,y} + dt\left( \frac{\partial E_z}{\partial x}\right)$$
$$ B_{\rm{new},z} = B_{t,z} - dt\left( \frac{\partial E_y}{\partial x}\right)$$

To test I took: 
$$ E_x = \sin(2 \pi x)$$
$$ E_y = \cos(2 \pi x)$$
$$ E_z = \sin(4 \pi x)$$

then 
$$ B_{\rm{new},x} = B_{t,x} $$

$$ B_{\rm{new},y} = B_{t,y} + dt\left(4 \pi \cos(4 \pi x) \right)$$

$$ B_{\rm{new},z} = B_{t,z} - dt\left( 2\pi \sin(2 \pi x) \right)$$

```python
dx = 0.02
dt = 0.01

x_primal = np.arange(len(Bx)) * dx
x_dual = np.arange(len(By)) * dx + 0.5 * dx

Bx_init = 0
By_init = 0
Bz_init = 0

Bx_theoretical = Bx_init + np.zeros(len(x_primal))
By_theoretical = By_init + dt*(4*np.pi*np.cos(4*np.pi*x_dual))
Bz_theoretical = Bz_init + dt*(2*np.pi*np.sin(2*np.pi*x_dual))

plt.figure(figsize=(8,5))
# Bug Fix: Make sure to not plot ghost cells.
plt.plot(x_primal[1:-1], Bx[1:-1], 'g-', label="Bx")
plt.plot(x_dual[1:-1], By[1:-1], 'b-', label="By")
plt.plot(x_dual[1:-1], Bz[1:-1], 'r-', label="Bz")

plt.plot(x_primal[1:-1], Bx_theoretical[1:-1], color = 'black', linestyle = '--')
plt.plot(x_dual[1:-1], By_theoretical[1:-1], 'k--')
plt.plot(x_dual[1:-1], Bz_theoretical[1:-1], 'k--')


plt.xlabel("x")
plt.ylabel("field")
plt.title("Faraday Test")
plt.legend()
plt.show()

# Bnew_y(ix) = By(ix) + m_dt* (Ez(ix) - Ez(ix-1))/ (dx);
# Bnew_z(ix) = Bz(ix) - m_dt* (Ey(ix) - Ey(ix-1))/ (dx);
```

```python
Ex = np.sin(2*np.pi*x_primal)
Ey = np.cos(2*np.pi*x_dual)
Ez = np.sin(4*np.pi*x_dual)

plt.plot(x_primal, Ex, label="Ex")
plt.plot(x_dual, Ey, label="Ey")
plt.plot(x_dual, Ez, label="Ez")
plt.legend()
```
