```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

uniform_bz = h5py.File("/home/debian/build/tests/boris/uniform_bz.h5")
drift_bz = h5py.File("/home/debian/build/tests/boris/drift_ey.h5")

x  = uniform_bz["x"][:]
vx = uniform_bz["vx"][:]
vy = uniform_bz["vy"][:]
vz = uniform_bz["vz"][:]
```

```python
plt.scatter(vx,vy)
plt.xlabel(r"$V_x$")
plt.ylabel(r"$V_y$")
plt.title("x-y Velocity Space")
plt.show()
```

The trajectory is a perfect circle in the (x,y)-plane.
The speed stays constant (no electric field, so kinetic energy conserved).

```python
dt = 0.001
t = np.arange(len(x)) * dt
plt.plot(t,vx)
plt.xlabel("time")
plt.ylabel("vx(t)")
plt.title("Velocity in time")
plt.show()
```

From the Lorentz Force, we have:

$$ \frac{dv}{dt} = \frac{q}{m} (v \times B) $$

$v_x(t)$ and $v_y(t)$ oscillate sinusoidally with frequency $$\omega_c = \frac{qB}{m} = 3$$

and dt = 0.001, $\omega = \frac{2 \pi}{f}$

```python
plt.scatter(vx,vz)
plt.xlabel(r"$V_y$")
plt.ylabel(r"$V_z$")
plt.title("y-z Velocity Space")
plt.show()
```

```python
dt = 0.001
t = np.arange(len(x)) * dt
plt.plot(t, x, color = 'black')
plt.xlabel("time")
plt.ylabel("x(t)")
plt.title("Position x in time")
plt.show()
```

# Drift

initially, motion is purely along y, perpendicular to B. Now the electric field is not 0.

$$ \frac{dv}{dt} = \frac{q}{m} (E + v \times B) $$

This creates in the steady state an $E \times B$ drift:
$$ v_{E \times B} = \frac{E \times B}{B^2 }$$

for $E=(0,1,0)$, $B=(0,0,8)$ we have a drift in x $$ v_{E \times B} = 0.125 \hat{x}$$

```python
x_drift  = drift_bz["x"][:]
vx_drift = drift_bz["vx"][:]
vy_drift = drift_bz["vy"][:]
vz_drift = drift_bz["vz"][:]
```

```python
dt = 0.001
v_drift = 1/8
t = np.arange(len(x_drift)) * dt
plt.plot(t, x_drift)
plt.plot(t, v_drift * t + x_drift[200], label=r'$v_{\mathrm{drift}}$ =' + f'{v_drift}' )
plt.xlabel("time")
plt.ylabel("x(t)")
plt.title("Position x in time with drift")

plt.legend()
```
